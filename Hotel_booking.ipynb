## **1.What is the overall cancellation rate for hotel bookings?**

```python
# Calculate the cancellation rate
cancellation_rate = df['is_canceled'].mean() * 100
cancellation_rate
```

```python
# Create a pie chart to visualize the cancellation rate
plt.figure(figsize = (10,6))
labels = ['Not Canceled', 'Canceled']
sizes = [100 - cancellation_rate, cancellation_rate]
colors = ['lightskyblue', 'lightcoral']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Overall Cancellation Rate for Hotel Bookings')
plt.show()
```

## **2.How does the lead time (number of days between booking and arrival) correlate with the cancellation rate?**

```python
# Group the data by lead time and calculate the cancellation rate for each lead time
lead_time_cancellation_rate = df.groupby('lead_time')['is_canceled'].mean()

# Create a scatter plot to visualize the correlation between lead time and cancellation rate
plt.figure(figsize = (10,6))
plt.scatter(lead_time_cancellation_rate.index, lead_time_cancellation_rate.values, alpha=0.5)
plt.xlabel('Lead Time (Days)')
plt.ylabel('Cancellation Rate')
plt.title('Correlation between Lead Time and Cancellation Rate')
plt.show()
```

## **3.What are the most common deposit types used for hotel bookings?**

```python
# Count the occurrences of each deposit type
deposit_counts = df['deposit_type'].value_counts()

# Define custom colors for the bar graph
colors = ['skyblue', 'lightgreen', 'orange']

# Create a color bar graph for the most common deposit types
plt.figure(figsize=(10, 6))
plt.bar(deposit_counts.index, deposit_counts.values, color=colors)
plt.xlabel('Deposit Type')
plt.ylabel('Frequency')
plt.title('Most Common Deposit Types for Hotel Bookings')
plt.show()
```

## **4.How does the average daily rate (ADR) vary across different customer types (e.g., transient, contract)?**

```python
# Group the data by customer type and calculate the average daily rate (ADR)
adr_by_customer_type = df.groupby('customer_type')['adr'].mean()

# Define custom colors for each customer type
colors = ['skyblue', 'orange', 'green', 'red']

# Create a bar chart with colored bars to visualize the ADR across different customer types
plt.figure(figsize = (10,6))
plt.bar(adr_by_customer_type.index, adr_by_customer_type.values, color=colors)
plt.xlabel('Customer Type')
plt.ylabel('Average Daily Rate (ADR)')
plt.title('ADR Variation Across Customer Types')
plt.show()
```

## **5.What is the distribution of special requests made by guests, and how does it impact the overall guest satisfaction?**

```python
# Count the occurrences of each special request
special_requests_counts = df['total_of_special_requests'].value_counts().sort_index()
special_requests_counts
```

```python
# Create a color palette for the bar chart
colors = plt.cm.viridis_r(range(len(special_requests_counts)))

# Create a bar chart to visualize the distribution of special requests
plt.figure(figsize=(10, 6))
plt.bar(special_requests_counts.index, special_requests_counts.values, color=colors)
plt.xlabel('Number of Special Requests')
plt.ylabel('Frequency')
plt.title('Distribution of Special Requests')
plt.show()
```

## **6.Are there specific months or seasons when hotel bookings are consistently higher or lower?**

```python
arrival_date_month = df['arrival_date_month']
arrival_date_month
```

```python
# Fetch the unique values of the 'arrival_date_month' column
unique_months = df['arrival_date_month'].unique()

# Print the unique months
print(unique_months)
```

```python
import seaborn as sns
# Extract the relevant columns for analysis (e.g., arrival_date_month and the count of bookings)
booking_counts = df['arrival_date_month'].value_counts().sort_index()

# Define the months and their corresponding numerical values
months = ['July', 'August', 'September', 'October', 'November', 'December', 'January',
          'February', 'March', 'April', 'May', 'June']

# Create a dictionary to store the average number of bookings for each month
average_bookings = {}

# Calculate the average number of bookings for each month
for month in months:
    month_bookings = booking_counts.loc[month]
    average_bookings[month] = month_bookings.mean()

# Sort the average bookings dictionary by month numerical value
sorted_average_bookings = dict(sorted(average_bookings.items(), key=lambda x: months.index(x[0])))

# Create a colorful horizontal bar plot to visualize the average number of bookings for each month
plt.figure(figsize=(10, 6))
colors = sns.color_palette('viridis', len(sorted_average_bookings))
sns.barplot(x=list(sorted_average_bookings.values()), y=list(sorted_average_bookings.keys()), palette=colors)
plt.xlabel('Average Number of Bookings')
plt.ylabel('Month')
plt.title('Average Number of Bookings by Month')
plt.show()
```

# Based on the analysis of the hotel booking dataset, the following conclusions can be drawn:

1. Cancellation Rate: The overall cancellation rate for hotel bookings is moderate, indicating the need for strategies to minimize cancellations and optimize revenue.

2. Lead Time Impact: There is a positive correlation between lead time and cancellation rate, suggesting that longer lead times may increase the likelihood of cancellations. Managing pricing and policies based on lead time can help improve booking conversion.

3. Deposit Types: The most common deposit types used for hotel bookings are "No Deposit" and "Non-Refundable." Offering flexible deposit options can cater to different guest preferences and increase booking commitment.

4. ADR by Customer Types: Average Daily Rate (ADR) varies across different customer types, with transient customers generally having higher rates. Targeting specific customer segments with tailored marketing and pricing strategies can maximize revenue.

5. Special Requests and Guest Satisfaction: The distribution of special requests made by guests varies, and their impact on overall guest satisfaction is mixed. Personalizing services based on guest preferences can enhance satisfaction and foster guest loyalty.

6. Seasonal Patterns: Hotel bookings exhibit seasonal fluctuations, with specific months or seasons experiencing consistently higher or lower demand. Aligning marketing efforts, pricing strategies, and resource allocation with seasonal patterns can optimize revenue and occupancy rates.

Overall, these insights highlight the importance of understanding guest preferences, optimizing pricing strategies, and providing exceptional guest experiences to drive booking performance and maximize revenue in the hotel industry.
